## Calculating gpd change for Arab World

```python
data = df_arab.values
```

```python
data
```

```python
gdp_change = [0]

for i in range(1, len(data)):

    prev = data[i-1][3]       # previous year gdp
    curr = data[i][3]          # current year gdp

    gdp_change.append(round(((curr - prev) / prev * 100),2)) 
```

```python
gdp_change
```

```python
# assigning new column for gpd change
df_arab.assign(GPD = gdp_change)
```

```python
len(gdp_change)
```

GDP growth for all the countries

```python
final_gdp_data = []

for country in df['Country Name'].unique():
    df_country = df[df['Country Name'] == country]

    data = df_country.values
    gdp_change = [0]  # first year has no previous value

    for i in range(1, len(data)):
        prev = data[i-1][3]
        curr = data[i][3]

        gdp_change.append(round(((curr - prev)/prev)*100, 2))
    
    df_country = df_country.assign(GDP = gdp_change)
    final_gdp_data.append(df_country)
```

```python
# final_gdp_data
```

```python
df = pd.concat(final_gdp_data, ignore_index = True )
```
